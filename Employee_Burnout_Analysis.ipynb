```python
# Load the dataset
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/Dataset(AI_Project)/Employee_burnout_analysis-AI_2.csv')
data
```

```python
# Check for missing values
print("Missing values before filling:")
data.isnull().sum()
```

```python
# Data preprocessing
# Fill missing values
data['Resource Allocation'].fillna(data['Resource Allocation'].mean(), inplace=True)
data['Mental Fatigue Score'].fillna(data['Mental Fatigue Score'].mean(), inplace=True)
data['Burn Rate'].fillna(data['Burn Rate'].median(), inplace=True)
```

```python
print("\nMissing values after filling:")
print(data.isnull().sum())
```

```python
data.head()
```

```python
data.tail()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# Encode categorical variables
categorical_cols = ['Gender', 'Company Type', 'WFH Setup Available']
label_encoders = {col: LabelEncoder() for col in categorical_cols}
for col in categorical_cols:
    data[col] = label_encoders[col].fit_transform(data[col])
```

```python
# Prepare features and target
X = data.drop(columns=['Employee ID', 'Date of Joining', 'Burn Rate'])
y = data['Burn Rate']
```

```python
# Scale numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
# Train a Random Forest Regressor
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)
```

```python
# Print evaluation metrics
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"RÂ² Score: {r2:.4f}")
```

```python
# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Feature Importance
importances = model.feature_importances_
feature_names = X.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances}).sort_values(by='Importance', ascending=False)
```

```python
# Plot Feature Importance
plt.figure(figsize=(10, 6))
sns.barplot(data=feature_importance_df, x='Importance', y='Feature', palette='viridis')
plt.title('Feature Importance')
plt.show()
```

```python
# Prediction vs. Actual Values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5, edgecolors='k', label='Predicted vs. Actual')
plt.plot([0, 1], [0, 1], '--r', label='Perfect Prediction')
plt.title('Prediction vs. Actual Burn Rate')
plt.xlabel('Actual Burn Rate')
plt.ylabel('Predicted Burn Rate')
plt.legend()
plt.show()
```

```python
# Prediction with User Input
def predict_burnout():
    print("\nEnter the following details for prediction:")
    gender = input("Gender (Male/Female): ")
    company_type = input("Company Type (Product/Service): ")
    wfh = input("WFH Setup Available (Yes/No): ")
    designation = int(input("Designation (1 to 5): "))
    resource_allocation = float(input("Resource Allocation (1 to 10): "))
    mental_fatigue_score = float(input("Mental Fatigue Score (0 to 10): "))

    # Encode user input
    user_data = pd.DataFrame({
        'Gender': [label_encoders['Gender'].transform([gender])[0]],
        'Company Type': [label_encoders['Company Type'].transform([company_type])[0]],
        'WFH Setup Available': [label_encoders['WFH Setup Available'].transform([wfh])[0]],
        'Designation': [designation],
        'Resource Allocation': [resource_allocation],
        'Mental Fatigue Score': [mental_fatigue_score]
    })

    # Scale user input
    user_scaled = scaler.transform(user_data)

    # Predict Burn Rate
    burn_rate = model.predict(user_scaled)[0]
    print(f"\nPredicted Burn Rate: {burn_rate:.4f}")
```

```python
# Call prediction function
predict_burnout()
```
